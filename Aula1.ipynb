```python
#Passo 1: Entrar no sistema da empresa (link do drive)
#Passo 2: Navegar no sistema e enconrtar a base de vendas (entrar na pasata exportar)
#Passo 3: Fazer o download da base de vendas
#Passo 4: Importar a base de vendas para Python
#Passo 5: Calcular os indicadores da empresa
#Passo 6: Enviar e-mail para a diretoria com os indicadores
```

```python
import pyautogui
import pyperclip
import time
import os

text = 'https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing'


pyautogui.PAUSE = 1

pyautogui.hotkey("ctrl", "t")
#pyperclip.copy("https://drive.google.com/drive/folders/149xknr9JvrlEnhNWO49zPcw0PW5icxga?usp=sharing") Nã funcionou 
os.system("echo '{}' | xclip -selection clipboard".format(text))
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")

time.sleep(5)

pyautogui.click(x=633, y=327, clicks=2)
time.sleep(2)


pyautogui.click(x=455, y=318)
pyautogui.click(x=1159, y=225)
pyautogui.click(x=984, y=632)
time.sleep(5)



```

### Vamos agora ler o arquivo baixado para pegar os indicadores

- Faturamento
- Quantidade de Produtos



```python
import pandas as pd

tabela = pd.read_excel(r"/home/danielle/Downloads/Vendas - Dez.xlsx")
display(tabela)
```

```python
faturamento = tabela["Valor Final"].sum()
print(faturamento)
quantidade = tabela["Quantidade"].sum()
print(quantidade)
```

### Vamos agora enviar um e-mail pelo gmail

```python
# abrir aba
pyautogui.hotkey("ctrl", "t")

# entrar no link do email - https://mail.google.com/mail/u/0/#inbox
text = "https://mail.google.com/mail/u/0/#inbox"
os.system("echo '{}' | xclip -selection clipboard".format(text))
pyautogui.hotkey("ctrl", "v")
pyautogui.press("enter")
time.sleep(5)

# clicar no botão escrever
pyautogui.click(x=171, y=234)

# preencher as informações do e-mail
pyautogui.write("danisamon@gmail.com")
pyautogui.press("tab") # selecionar o email

pyautogui.press("tab") # pular para o campo de assunto
os.system("echo '{}' | xclip -selection clipboard".format("Relatório de Vendas"))
pyautogui.hotkey("ctrl", "v")

pyautogui.press("tab") # pular para o campo de corpo do email

texto = f"""
Prezados,

Segue relatório de vendas.
Faturamento: R${faturamento:,.2f}
Quantidade de produtos vendidos: {quantidade:,}

Qualquer dúvida estou à disposição.

"""


os.system("echo '{}' | xclip -selection clipboard".format(texto))
pyautogui.hotkey("ctrl", "v")

# enviar o e-mail
pyautogui.hotkey("ctrl", "enter")
```
